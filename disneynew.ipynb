```python
sns.set(font_scale=1.1)

plt.figure(figsize=(10,8))
sns.countplot(x='type',data=df,palette='cool')
plt.title("Product Types", size=15)
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year #might delete dateadded colm
```

```python
plt.figure(figsize=(10,8))
pd.value_counts(df['rating']).plot.bar()
plt.title('Ratings for Content')
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df,x='year_added', hue='rating')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df,x='rating', hue='type')
```

```python
plt.figure(figsize=(10,8))
df["director"].value_counts().head(10).plot(kind='barh')
plt.title("Directors with Most Movies", size=15)
plt.show()
```
